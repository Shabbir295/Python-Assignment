3. Write a program in which use all the operators we can use in Python

```python
num1 = 10
num2 = 8
```

```python
## Arithmetic Operators
print (num1 + num2) 
print (num1 - num2)
print (num1 * num2)
print (num1 / num2)
print (num1 % num2)
print (num1 ** num2)
print (num1 // num2)
```

```python
#Assignment Operators
print (num1 == 10)
print (num1 != 10)
print (num1 > 10)
print (num1 < 10)
print (num1 >= 10)
print (num1 <= 10)
```

```python
#Comparision Operators
num1 += num2
print (num1)
num1 -= num2 
print (num1)
num1 *= num2
print (num1)
num1 /= num2
print (num1)
num1 %= num2
print (num1)
num1 **= num2
print (num1)
num1 //= num2
print (num1)
```

```python
#Logical Operators
print((num1 > num2) and (num1 < 20))
print((num1 > num2) or (num1 < num2))
print(not (num1 == num2))
```
